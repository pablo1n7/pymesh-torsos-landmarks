```python
import numpy as np
import pymesh
from glob import glob
from sklearn import preprocessing
```

```python
fs_mesh = np.array(sorted(glob("./out/torsos/*.ply")))
```

```python
inputer = preprocessing.Imputer(missing_values=0,strategy="mean", axis=0)
```

```python
magic_number = 58852
```

```python
values = []
for mesh_f in fs_mesh:
    mesh = pymesh.load_mesh(mesh_f)
    temp_array  = np.zeros((magic_number,3))
    #print np.shape(mesh.vertices[0:magic_number])
    temp_array[:mesh.vertices.shape[0]] = mesh.vertices
    values.append(temp_array)
```

```python
shapes_values = []
for i in values:
    shapes_values.append(np.shape(i)[0])
np.min(shapes_values)
```

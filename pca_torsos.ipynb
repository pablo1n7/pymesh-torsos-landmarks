```python
# 20 y 25 como normales
#  Menos de 20 bajo peso
#  25 a 30 sobre peso
#  MÃ¡s de 30 obesidad

def bmi_cal(bmi):
    if 18.5 > bmi:
        return "underfat"
    if 18.5 <= bmi < 25:
        return "healthy"
    if 25 <= bmi < 30:
        return "overfat"
    if 30 <= bmi:
        return "obese"
```
